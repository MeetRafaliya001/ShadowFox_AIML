#House price prediction model

Here i am using linear regression method because data is continues.

```python
#step 1: import library
import pandas as pd
```

```python
#step 2: read database
df = pd.read_csv('https://drive.usercontent.google.com/download?id=1HCH268MvclbWQtW41YtDMhN-yxGvXZKQ&export=download&confirm=t&uuid=2c0b753e-a294-4b9f-bb47-f29e4e94e53c')
```

```python
#step 3: Define y and x
df.columns
```

```python
y = df['MEDV']
X = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
```

```python
#step 4: Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=2529)
df.shape,X_train.shape,X_test.shape
```

```python
506*.8
```

```python
# step 5: select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# step 6: train model
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)
model.fit(X_train,y_train)
```

```python
# step 7: prediction
imputer = SimpleImputer(strategy='mean')
X_test = imputer.fit_transform(X_test)
y_pred = model.predict(X_test)
```

```python
# step 8: evaluate
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```
